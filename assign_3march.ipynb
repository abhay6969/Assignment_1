```python
"""Answer 1:
Scatter Plot: A scatter plot is a two-dimensional plot that displays the relationship
between two variables.

Line Plot: A line plot displays the relationship between two or more continuous 
variables . This plot is useful for showing trends and changes in the data over time .
It can be used for a range of applications, such as plotting time series data or
visualizing changes in a continuous variable over different categories.

Histogram: A histogram is a plot that shows the distribution of a single
variable. This plot is useful for understanding the range of the data and things whether
there are any outliers.Histogram allows us to easily customize the bin size, color, and
otheraspects of the plot to highlight important features of the data.

Bar Plot: A plot that displays the mean or count of a categorical variable with bars. 
It can be used to compare the values of the categorical variable across different groups.

Pair Plot: A plot that displays the pairwise relationships between multiple variables in
a dataset. It can be used to identify patterns and correlations between variables.
"""
```

```python
#Scatter Plot
import seaborn as sns
tips = sns.load_dataset('tips')
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
```

```python
#Line Plot
tips = sns.load_dataset('tips')
sns.lineplot(x="total_bill", y="tip", hue="sex", data=tips)
```

```python
#Histogram Plot
tips = sns.load_dataset('tips')
sns.histplot(x="total_bill", data=tips, bins=20)
```

```python
#Bar Plot
tips = sns.load_dataset('tips')
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
#Pair Plot
tips = sns.load_dataset('tips')
sns.pairplot(tips, hue='sex', palette='husl')
```

```python
#Answer 2:
fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
#Answer 3:
titanic = sns.load_dataset("titanic")
sns.boxplot(x='pclass', y='age', data=titanic)
```

```python
sns.boxplot(x='pclass', y='fare', data=titanic)
```

```python
diamonds = sns.load_dataset("diamonds")

sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack")
```

```python
#Answer 4:
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue="species")
```

```python
#Answer 5:
flights = sns.load_dataset("flights")
flights_matrix = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(data=flights_matrix)
```
