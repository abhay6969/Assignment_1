```python
"""Answer 1:
Matplotlib is a Python library for creating interactive
visualizations in Python. It is a widely used plotting library for data visualization. Matplotlib
provides a variety of visualization tools, such as  line plot and scatter plot.

Matplotlib is used to:
Visualize data in a clear and concise way and Identify patterns and relationships in 
data.

Five plots that can be plotted using the Pyplot module of Matplotlib are:
1. Line Plot - used to show changes over time
2. Scatter Plot - used to show the relationship between two variables
3. Bar Plot - used to show comparisons between different categories 
4. Histogram - used to show the distribution of a  variable
5. Pie Chart - used to show the proportion of different categories in a dataset."""
```

```python
"""Answer 2:
A scatter plot is a type of plot used to display the relationship between two variables.
here , two variables are plotted along two axis and the resulting pattern shows the
corelation present between them"""
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

```python
"""Answer 3:
The subplot() function in Matplotlib is used to create multiple plots in the same figure.
This function is useful when we want to compare multiple plots"""

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

plt.subplot(4, 4, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(4, 4, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(4, 4, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(4, 4, 4)
plt.plot(x4, y4)
plt.title('Line 4')

plt.show()
```

```python
"""Answer 4:
A bar plot is a type of chart that represents categorical data with rectangular bars 
A bar plot is used to show comparisons between different categories of data.It can be
either vertical or horizontal."""
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.figure(figsize=(5, 5))
plt.subplot(2,2,1)
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Profit by Company (Vertical)')
plt.show()

# Horizontal bar plot
plt.subplot(2,2,2)
plt.barh(company, profit)
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Profit by Company (Horizontal)')
plt.show()
```
