```python
"""Answer 1:
We can create a Bokeh plot using Python by importing figure and show function from the bokeh.plotting library"""

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers

plot = figure(title="Iris Data", x_axis_label="Petal Length", y_axis_label="Petal Width")
plot.circle(flowers["petal_length"], flowers["petal_width"], size=10)

output_file("iris.html")
show(plot)
```

```python
"""Answer 2:
Glyphs in Bokeh are markers that represent data points, such as circles, squares, triangles, and some other shapes.
 we can add glyphs to a Bokeh plot using the glyph method of a Figure object"""

import numpy as np
x = np.random.normal(size=100)
y = np.random.normal(size=100)

# create a figure object
plot = figure(title="Random Data", x_axis_label="X", y_axis_label="Y")

# add a circle glyph
plot.circle(x, y, size=10, color="blue")
plot.triangle(x, y, size=5, color="red")

# specify an output file
output_file("random.html")

# show the plot
show(plot)
```

```python
#Answer 3:
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource

from bokeh.sampledata.iris import flowers
source = ColumnDataSource(flowers)

plot = figure(title="Iris Data", x_axis_label="Petal Length", y_axis_label="Petal Width")
plot.circle("petal_length", "petal_width",legend_field = 'species', color = 'red',size=10,source = source)

plot.title.text_color = "red"
plot.xaxis.axis_label_text_color = "green"
plot.yaxis.axis_label_text_color = "purple"
plot.legend.location = "top_left"

output_file("iris.html")

show(plot)
```
